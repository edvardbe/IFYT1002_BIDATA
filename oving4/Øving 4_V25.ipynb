## Oppgave 1

Kula i et flipperspill skytes ut fra en fjærbelastet avtrekker. Spilleren trekker i fjæra slik at den presses sammen en avstand $x=7,0 \textrm{ cm}$. Massen til kula er $80 \textrm{ g}$, og flipperspillet har en helningsvinkel $\theta=15^{\circ}$. Kula sklir friksjonsfritt mot underlaget. Se figuren under.

<center><img src="bilder/pinball_fjær.png" alt="" style="width: 500px;"/></center>

Hvor stor må fjærkonstanten $k$ til fjæra være dersom kula akkurat skal nå toppen av flipperspillet, som ligger en avstand $s=78 \textrm{ cm}$ fra punktet der kula forlater fjæra (i punktet der fjæra er slapp)?



### Løsning
Når kulen forlater fjæren ved $x=0$ så er all den potensielle energien i fjæren, $U_f$, gjort om til kinetisk energi, $K$, og noe potensiell energi i tyngdefeltet, $U_g$. Alstå er:
\
\
$U_f = K + U_g$
\
\
$\frac{1}{2} k x^2 =\frac{1}{2} m v_0^2 + mgh = m (\frac{1}{2} v_0^2 + gh)$
\
\
Dermed trenger vi utskytningshastigheten, $v_0$, som vi kan finne ved å tenke oss at det kun er gravitasjonen som virker på ballen etter at den har forlatt fjæren og vi antar en hastighet på, $v=0$, når ballen har nådd en strekning på $s=0.78m$. Deretter kan vi skrive om formelen for strekning, den som ikke er avhengig av tid, for å finne $v_0$.
\
\
$v_0 = \sqrt{v^2 - 2a_xs}$
\
\
Akselerasjonen $a_x$ kan vi finne med trigonometri der $G_x = G \cdot sin(\theta-180\degree) \rightarrow a_x = \frac{G_x}{m}$. Vinkelen er $\alpha = \theta - 180 \degree$ fordi akselerasjonen i x-retning virker negativt.
\
\
$a_x = g sin (\theta-180\degree) = 9.81m/s^2 \cdot sin(\frac{\pi}{12} - \pi) = -2.54m/s^2 ≈ -2.5m/s^2$
\
\
$v_0 = \sqrt{v^2 - 2a_xs} = \sqrt{0 - 2 \cdot -2.5m/s^2 \cdot 0.78m} = 1.98m/s ≈ 2.0m/s$
\
\
$\frac{1}{2} k x^2 = 0.080kg (\frac{1}{2} (2.0m/s)^2 + 9.81m/s^2 \cdot sin(\frac{\pi}{12}) \cdot 0.070m)= 1.74 J ≈ 1.7 J$
\
\
$k = \frac{(K + U_g) \cdot 2}{x^2} =\frac{0.17 J \cdot 2}{(0.070m)^2} = 69 N/m$

Eksempel: for ``` a=np.array([0.1,2.4,4.0,9.4])```  vil ``` np.argmax(a>3)```  returnere ``` 2``` , dvs. ``` a[2]=4.0```  er det første elementet som oppfyller a>3. I denne oppgaven: Dersom terminalfarten er ``` v_terminal```, og ```v_liste``` er en array med fartsverdiene, dvs. $v(t)$, og man vil finne ut hvilken $t$-verdi som tilsvarer 98 % av terminalfarten, blir syntaksen ```np.argmax(v>0.98*v_terminal)``` ("første/laveste indeks som oppfyller at farten er **større** enn 98 % av terminalfarten").
````python
#Rutiner for simulering av vertikalt fall med luftmotstand. I dette eksemplet er
# positiv retning nedover.

import numpy as np
import matplotlib.pyplot as plt

# Globale konstanter
m=70 #Legemets masse i kg
g=9.81 #Tyngdeakselerasjonen i m/s^2

def drag_kraft(A,C,v):
    rho=1.29 
    k=0.5*rho*A*C
    #? Her må du legge inn riktig kode for at funksjonen skal returnere 
    #luftmotstanden (drag) i newton
    return k*v**2

def dXdt(X,A, C):
   """Funksjonen dXdt beregner høyresiden f(X) i differensiallikningssystemet; dX/dt=f(X).
   Input:
   X:  X=[y,v] en vektor som inneholder posisjon y og (vertikal) fart v. Med positiv retning
   nedover er v > 0 og y < 0    for et legeme som faller vertikalt mot bakken, der y = 0.
   
   Output:
   [dydt,dvdt]: Array med nye verdier for hastighet (dydt) og akselerasjon (dvdt)
   """
    #Koeffisienter for luftmotstand
   y , v =X        #Koordinater y og v hentes fra inndatavektor X
   f=drag_kraft(A,C,v)   #Luftmotstand i N
   dydt=v          #Sammenhengen mellom y og v er at v = dy/dt
   dvdt=-f/m+g     #Akselerasjonen a=dv/dt, fra Newtons 2. lov
   return np.array([dydt,dvdt])

def euler(t0,y0,v0,dt, A, C):
    """Funksjon som bruker Eulers metode til å løse et system av differensiallikninger,
    dX/dt = f(X), der X =[y,v] er
    en vektor som inneholder posisjons- og hastighetsvariable. 
    Input: 
    t0: Starttid [s]
    y0: Startverdi for y [m]
    v0: Vertikal startfart [m/s]
    dt: Tidssteg [s]

    Output:
    t_liste: array med t-verdier,[t0,...tn]
    y_liste: array med y-verdier, [y0,...,yn]
    v_liste: array med v-verdier, [v0,...,vn]
    """

    X0=np.array([y0,v0]) #X0 er en vektor med posisjon og fart ved t=t0
    t_liste=[0.0]# Liste med t-verdier
    y_liste=[y0]# liste med y-verdier
    v_liste=[v0] # liste med v-verdier
    X=X0  # initierer loop
    t=t0
    y=y0
    while y<=0: #Loop kjøres inntil legemet treffer bakken; pos. retning nedover gir y0 < 0
        Xn=X+dt*dXdt(X, A, C) #Beregner neste steg Xn i Euler-metoden
        y=Xn[0] #Henter ut y-koordinat fra array
        v=Xn[1] #Henter ut fart v fra array
        t_liste.append(t) # t-verdi legges til liste
        y_liste.append(y)# y-verdi legges til liste
        v_liste.append(v)# v-verdi legges til liste
        t=t+dt #Ny tidsverdi    
        X=Xn #Ny verdi for X 
    return t_liste,y_liste,v_liste
````

````python
#Eksempel på bruk av rutinene
#Initialiserer variable 
t0=0.0 #t = 0 i startpunktet
v0=0 #Startfart
y0=-2000 #Med positiv retning nedover, er y-verdier negative over bakken. 
#Her starter fallet 2,0 km over bakken.
dt=0.1 #Tidssteg
A=0.17 #Frontareal
C=0.7 #Drag-koeffisient
A_1=1.0 #Frontareal
C_1=1.0 #Drag-koeffisient

#Bruker Eulers metode til å generere sammenhengende verdier for t, y og v
t_liste,y_liste,v_liste=euler(t0,y0,v0,dt, A, C)
t1_liste,y1_liste,v1_liste=euler(t0,y0,v0,dt, A_1, C_1)

#Plotter fartsgraf v(t)
plt.figure(figsize = (10, 8))
#Diagrammet angir hvilke verdier for A og C som er tilhører grafen
plt.plot(t_liste,y_liste,color="red",label='A = '+str(A)+", C = "+str(C)) 
plt.plot(t1_liste,y1_liste,color="blue",label='A = '+str(A_1)+", C = "+str(C_1))
plt.xlabel("Tid [s]")
plt.ylabel("Fart [m/s]")
plt.legend()
plt.show()
v_aterminal = 340/3.6
v_bterminal = 117/3.6
v_liste = np.array(v_liste)
v1_liste = np.array(v1_liste)
indeks_a = np.argmax(v_liste > 0.98*v_aterminal)
indeks_b = np.argmax(v1_liste > 0.98*v_bterminal)
print("Indeksen for A for 98% terminal fart er: ", indeks_a)
print("Tiden for A for 98% terminal fart er: ", t_liste[indeks_a])
print("Strekning for A for 98% terminal fart er: ", 2000 + y_liste[indeks_a])

print("Indeksen for B for 98% terminal fart er: ", indeks_b)
print("Tiden for B for 98% terminal fart er: ", t1_liste[indeks_b])
print("Strekning for B for 98% terminal fart er: ", 2000 + y1_liste[indeks_b])
````
